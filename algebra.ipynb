## Table's extension by powerset with various dimensions and number inputs and outputs

n - input dimension

m - output dimension

p - number of inputs

q - number of outputs

# with null $\varnothing$

$\text(p + q)D$ table

$\underbrace{2^n\times{\ldots}\times2^n}_{p-times}\times \underbrace{2^m\times{\ldots}\times2^m}_{q-times}$

or

$\text(p + 1)D$ table

$\underbrace{2^n\times{\ldots}\times2^n}_{p-times}\times 2^{qm}$

or

$\text(1 + q)D$ table

$ 2^{pm}\times\underbrace{2^m\times{\ldots}\times2^m}_{q-times}$

or

$\text2D$ table

$2^{pn}\times 2^{qm}$

relations (functions)

$2^{pn}2^{qm}=2^{pn+qm}$

# without null $\varnothing$

$\left({p + q}\right)D$ table 

$ \underbrace{(2^n-1)\times{\ldots}\times(2^n-1)}_{p-times}\times \underbrace{(2^m-1)\times{\ldots}\times(2^m-1)}_{q-times}$

relations (functions)

$(2^n-1)^p(2^m-1)^q$

```
import math
n = 3  # input dim, e.g. RGB space (R, G, B)
m = 3  # output dim, e.g. RGB space (R, G, B)
p = 3  # n inputs, e.g. red (1, 0, 0), green (0, 1, 0) and blue (0, 0, 1) colors    
q = 2  # n outputs, arbitrary two colors: (R1, G1, B1) and (R2, G2, B2)
print('dim in: {}\ndim out: {}\nn in: {}\nn out: {}'.format(m, n, p, q))
print('with null')
i = 2**(p*n)
o = 2**(q*m)
f = 2**(p*n + q*m)
print('table (' + ' x '.join(['{}'.format(2**n) for _ in range(p)]) 
      + ') x (' + ' x '.join(['{}'.format(2**m) for _ in range(q)]) + ')')
print('function input types: {}'.format(i))
print('function output types: {}'.format(o))
print('types of functions: {}'.format(f))
print('without null')
i = (2**n - 1)**p
o = (2**m - 1)**q
f = (2**n - 1)**p*(2**m - 1)**q
print('table (' + ' x '.join(['{}'.format(2**n - 1) for _ in range(p)]) 
       + ') x (' + ' x '.join(['{}'.format(2**m - 1) for _ in range(q)]) + ')')
print('function input types: {}'.format(i))
print('function output types: {}'.format(o))
print('types of functions: {}'.format(f))
```
